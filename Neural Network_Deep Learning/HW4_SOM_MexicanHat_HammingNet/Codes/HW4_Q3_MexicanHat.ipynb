```python
import numpy as np
np.set_printoptions(formatter={'float': lambda x: "{0:0.2f}".format(x)})
# To plot pretty figures
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)
```

```python
def f_activation(x):
  f=np.zeros(len(x))
  for j in range(len(x)):
    if (2>x[j]>0):
      f[j]=x[j]
    elif (x[j]>=2) :
      f[j]=2
    else:
      f[j]=0
  return f
```

# Part 1


```python
X_arr=np.array([0.32,0.33,0.28,0.47,0.66,0.80,0.4,0.33,0.1,0.26]).reshape(-1,1)
```

```python
#params
# R1=0
# R2=infinite
t_max=3
C2=-0.1
C1=0.2
```

```python
X_old=X_arr.copy()
X_new=np.zeros(len(X_arr)).reshape(-1,1)
X_new1=np.zeros((t_max,len(X_arr)))
for j in range(t_max):
  for i in range(len(X_arr)):
    X_new[i]=C2*np.sum(X_old-X_old[i])+C1*X_old[i]
  # X_new=np.multiply(X_new,C2)
  X_new1[j,:]=f_activation(X_new)
  X_old=X_new1[j,:].copy()
```

```python
X_new1[-1,:]
```

```python
plt.figure(figsize=(11,5))
plt.plot(X_new1[0,:],'ro-',label='t=1')
plt.plot(X_new1[1,:],'bo-',label='t=2')
plt.plot(X_new1[2,:],'go-',label='t=3')
plt.legend()
plt.xlabel('indices')
plt.ylabel('Values')
plt.title('Mexican Hat during Steps')
```
